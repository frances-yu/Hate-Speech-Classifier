```python
import os
import numpy as np
import pandas as pd
```

```python
def loadCSV(filename):
    file = filename
    if '.csv' not in filename:
        file += '.csv'
    data = pd.read_csv(file, encoding = 'ISO-8859-1')
    return data
```

```python
data_1 = loadCSV('tweetset_1')
data_2 = loadCSV('tweetset_2')
```
